```python
# Import Libraries
import requests
import pandas as pd
import matplotlib.pyplot as plt

# Optional: Configure matplotlib for inline display in Jupyter
%matplotlib inline
```

## Understand the Open-Meteo API

- **API Endpoint**: `https://archive-api.open-meteo.com/v1/archive`
- **Parameters**:
  - `latitude`: Latitude of the location.
  - `longitude`: Longitude of the location.
  - `start_date`: Start date of the data (YYYY-MM-DD).
  - `end_date`: End date of the data (YYYY-MM-DD).
  - `hourly`: Comma-separated list of variables (e.g., temperature, precipitation).
  - `timezone`: Timezone of the data.

**Documentation**: [Open-Meteo API Docs](https://open-meteo.com/en/docs)

```python
# Step 1: Set Up Parameters

# Location coordinates for New York City
latitude = 40.7128
longitude = -74.0060

# Date range
start_date = '2023-01-01'
end_date = '2023-01-07'  # One week of data

# Variables to fetch
hourly_variables = ['temperature_2m', 'relativehumidity_2m', 'precipitation']
```

```python
# Step 2: Make the API Request

base_url = 'https://archive-api.open-meteo.com/v1/archive'

params = {
    'latitude': latitude,
    'longitude': longitude,
    'start_date': start_date,
    'end_date': end_date,
    'hourly': ','.join(hourly_variables),
    'timezone': 'America/New_York'
}

response = requests.get(base_url, params=params)
```

```python
# Check if the request was successful

if response.status_code == 200:
    print('Data fetched successfully!')
else:
    print(f'Failed to fetch data. Status code: {response.status_code}')
```

```python
# Step 3: Load Data into pandas DataFrame

data = response.json()
hourly_data = data['hourly']

df = pd.DataFrame(hourly_data)
df.head()
```

```python
# Step 4: Data Cleaning

# Convert 'time' column to datetime
df['time'] = pd.to_datetime(df['time'])

# Set 'time' as the index
df.set_index('time', inplace=True)

# Check for missing values
df.isnull().sum()
```

```python
# Handle missing values (if any)
df.fillna(method='ffill', inplace=True)  # Forward fill
```

```python
# Step 5: Exploratory Data Analysis

# Summary Statistics
df.describe()
```

```python
# Plot Temperature Over Time

plt.figure(figsize=(14, 6))
plt.plot(df.index, df['temperature_2m'], label='Temperature (C)')
plt.title('Temperature Over Time in New York City')
plt.xlabel('Date')
plt.ylabel('Temperature (C)')
plt.legend()
plt.show()
```

```python
# Plot Relative Humidity Over Time

plt.figure(figsize=(14, 6))
plt.plot(df.index, df['relativehumidity_2m'], color='orange', label='Relative Humidity (%)')
plt.title('Relative Humidity Over Time in New York City')
plt.xlabel('Date')
plt.ylabel('Relative Humidity (%)')
plt.legend()
plt.show()
```

```python
# Plot Precipitation Over Time

plt.figure(figsize=(14, 6))
plt.bar(df.index, df['precipitation'], label='Precipitation (mm)')
plt.title('Precipitation Over Time in New York City')
plt.xlabel('Date')
plt.ylabel('Precipitation (mm)')
plt.legend()
plt.show()
```

```python
# Step 6: Correlation Analysis

# Calculate correlation between temperature and humidity
correlation = df['temperature_2m'].corr(df['relativehumidity_2m'])
print(f'Correlation between Temperature and Relative Humidity: {correlation:.2f}')
```

```python
# Scatter Plot of Temperature vs. Relative Humidity

plt.figure(figsize=(8, 6))
plt.scatter(df['temperature_2m'], df['relativehumidity_2m'], alpha=0.5)
plt.title('Temperature vs. Relative Humidity')
plt.xlabel('Temperature (C)')
plt.ylabel('Relative Humidity (%)')
plt.show()
```

```python
# Step 7: Resample Data to Daily Averages

daily_avg = df.resample('D').mean()
daily_avg.head()
```

```python
# Plot Daily Average Temperature

plt.figure(figsize=(10, 5))
plt.plot(daily_avg.index, daily_avg['temperature_2m'], marker='o')
plt.title('Daily Average Temperature in New York City')
plt.xlabel('Date')
plt.ylabel('Temperature (C)')
plt.grid(True)
plt.show()
```

```python
# Step 8: Compare with Another Location (Los Angeles)

# Coordinates for Los Angeles
latitude_la = 34.0522
longitude_la = -118.2437

params_la = {
    'latitude': latitude_la,
    'longitude': longitude_la,
    'start_date': start_date,
    'end_date': end_date,
    'hourly': ','.join(hourly_variables),
    'timezone': 'America/Los_Angeles'
}

response_la = requests.get(base_url, params=params_la)

# Load and clean Los Angeles data
data_la = response_la.json()
hourly_data_la = data_la['hourly']
df_la = pd.DataFrame(hourly_data_la)

df_la['time'] = pd.to_datetime(df_la['time'])
df_la.set_index('time', inplace=True)
df_la.fillna(method='ffill', inplace=True)
```

```python
# Resample Los Angeles Data to Daily Averages

daily_avg_la = df_la.resample('D').mean()
```

```python
# Combine DataFrames for Comparison

combined_temp = pd.DataFrame({
    'New York': daily_avg['temperature_2m'],
    'Los Angeles': daily_avg_la['temperature_2m']
})
```

```python
# Plot Comparison of Daily Average Temperature

combined_temp.plot(kind='bar', figsize=(10, 6))
plt.title('Daily Average Temperature: New York vs Los Angeles')
plt.xlabel('Date')
plt.ylabel('Temperature (C)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Step 9: Advanced Analysis - Temperature Difference

# Calculate Temperature Difference
combined_temp['Temp Difference'] = combined_temp['Los Angeles'] - combined_temp['New York']
combined_temp
```

```python
# Step 10: Save Data to CSV (Optional)

# Save New York data
df.to_csv('new_york_weather.csv')

# Save Los Angeles data
df_la.to_csv('los_angeles_weather.csv')
```

```python
# Import Libraries
import requests
import pandas as pd
import matplotlib.pyplot as plt

# Configure matplotlib for inline display in Jupyter
%matplotlib inline
```

```python
# Set Up Parameters

#Define coordinates and timezone for London and Seattle
cities = [{'name':'London', 'latitude': 51.5074, 'longitude': -0.1278, 'timezone': 'Europe/London'},
          {'name':'Seattle', 'latitude': 47.6062, 'longitude': -122.3321, 'timezone': 'America/Los_Angeles'}]

# Date range
start_date = '2022-01-01'
end_date = '2022-12-31'  # One year of data
```

```python
# Make the API Request and create dataframe

base_url = 'https://archive-api.open-meteo.com/v1/archive'

cities_df = []

for city in cities:
    params = {
    'latitude': city['latitude'],
    'longitude': city['longitude'],
    'start_date': start_date,
    'end_date': end_date,
    'hourly': 'precipitation',
    'timezone': city['timezone']
    }
    
    response = requests.get(base_url, params=params)

    # Check if the request was successful
    if response.status_code == 200:
        print('Data fetched successfully!')
    else:
        print(f'Failed to fetch data. Status code: {response.status_code}')

    data = response.json()
    hourly_data = data['hourly']
    df = pd.DataFrame(hourly_data)
    
    # Rename precipitation column to include city name
    df.rename(columns={'precipitation': f"{city['name']}_precipitation"}, inplace=True)
    cities_df.append(df)
    
# Merge all cities on the 'time' column
final_df = cities_df[0]
for df in cities_df[1:]:
    final_df = final_df.merge(df, on='time', how='outer')

# Display first few rows
print(final_df.head())

```

```python
# Data Cleaning

df = final_df

# Convert 'time' column to datetime
df['time'] = pd.to_datetime(df['time'])

# Set 'time' as the index
df.set_index('time', inplace=True)

# Check for missing values
df.isnull().sum()
```

```python
# Handle missing values (if any)
df.fillna(method='ffill', inplace=True)  # Forward fill
```

```python
# Resample Data
df = df.resample('W').sum()
```

```python
# Exploratory Data Analysis

# Summary Statistics
df.describe()
```

```python
df.head()
```

```python
# Visualization: Plot Weekly Precipitation Over Time for London and Seattle

plt.figure(figsize=(14, 6))
plt.plot(df.index, df['London_precipitation'], label='London')
plt.plot(df.index, df['Seattle_precipitation'], label='Seattle')
plt.title('Weekly Precipitation Over Time in London and Seattle')
plt.xlabel('Date')
plt.ylabel('Precipitation (mm)')
plt.legend()
plt.show()
```

```python
# Determine which city had more rainfall overall and identify any interesting patterns.

# Plot total rainfall for the year in each city
plt.figure(figsize=(14, 6))
plt.bar(['London', 'Seattle'], [df['London_precipitation'].sum(), df['Seattle_precipitation'].sum()])
plt.title('Total Precipitation in London and Seattle')
plt.xlabel('City')
plt.ylabel('Precipitation (mm)')
plt.legend()
plt.show()
```

```python
# Calculate total 2022 rainfall in London and Seattle

print(df['London_precipitation'].sum())
print(df['Seattle_precipitation'].sum())
```

### Analysis

As we can see from the bar graph, Seattle had more total rainfall than London in 2022. Specifically, Seattle had about 1,131mm while London had about 663mm. Aside from total rainfall, there are some other interesting trends that can be observed from the graphs. For one, both cities saw a significant drop in rainfall during the summer months of July and August. Seattle's drop in rainfall continued through October, while London saw an increase starting in September. During January and March, Seattle's total rainfall was much higher than London's, accounting for the overall gap in yearly rainfall between the two cities.

**1. Key Measure of Success**

Success in this experiment would mean an increase in cross-city driver share. Cross-city driver share is defined as the percentage of drivers who complete trips in both Gotham and Metropolis during the experiment window.

This is a good measure of success because it would directly show that more drivers are accepting trips in both cities any given week. 


**2. Experimental Design**

**Implementation:** Run a 4–6 week randomized controlled trial among eligible driver partners. Randomly assign drivers to either a control group with no toll reimbursement or a treatment group that receives full toll reimbursement. Keep assignment fixed throughout the experiment and measure behavior at the driver level over the test period.

**Statistical tests:** Use a two-sample test for proportions to compare cross-city driver share between the treatment and control groups. Since the outcome is whether or not a driver completed trips in both cities during the experiment window, this can be treated as a binary variable at the driver level. The test would show whether the difference in cross-city driver share between the two groups is statistically significant or likely due to random chance.

**Interpretation and recommendations:** If the treatment group shows a statistically significant increase in cross-city driver share, that would suggest that reimbursing toll costs is effective in encouraging more drivers to serve both cities. In that case, I would recommend considering a broader rollout of the policy. 

I would also want to look at the size of the increase and compare it to the cost of the reimbursements, since statistical significance alone does not guarantee that the policy makes sense from a business perspective. If there is no statistically significant difference, I would not recommend a full rollout based on this experiment alone. That result could mean the reimbursement is not a strong enough incentive, or that other barriers besides toll cost are keeping drivers from working in both cities.
